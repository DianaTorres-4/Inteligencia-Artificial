```python
import numpy as np

# Datos XOR
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])
```

Ingresamos los datos y valores de nuestra tabla de XOR en la primera linea de X los ingresamos en forma de lista los valores de X1 y X2 y lo que tenemos como la variable y es lo que se pretende tener de salida de la tabla de XOR

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

modelo = Sequential([
    Dense(4, activation='relu', input_dim=2),
    Dense(1, activation='sigmoid')
])
```

Despues vamos a definir el modelo de la red neuronal, por sonsiguiente utilizaremos el sequential que es para definir el modelo de la red neuronal añadiremos al mismo modelo capas de entrada y capas que esten conectadas entre si y tambien agregaremos capas de salida con la activacion sigmoid que esto nos sirve para clasificar la respuesta en 1 y 0  

```python
from tensorflow.keras.optimizers import Adam

# Tasa de aprendizaje deseada
learning_rate = 0.01
adam_optimizer = Adam(learning_rate=learning_rate)
```

El optimizador Adam es el que ajusta los pesos de las neuronas mientras el modelo aprende o no segun sea el caso y se define la tasa de aprendizaje de 0.01

```python
modelo.compile(
    optimizer=adam_optimizer,
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

El código establece el optimizador (Adam), la función de pérdida (Entropía Cruzada Binaria) y la métrica que el modelo usará para aprender de los datos.

```python
history = modelo.fit(
    X, y,
    epochs=50,
    verbose=1,
)
```

toma el modelo fit, y trata de alimentarlo con los datos X, y, lo que lo  hace aprender y ajustar sus parámetros durante 50 épocas, guardando el historial de rendimiento.


```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.title('Evolución de la pérdida (Loss)')
plt.xlabel('Época')
plt.ylabel('Binary Cross-Entropy')
plt.show()
```

Se utiliza la librería Matplotlib "plt", que es la herramienta para crear gráficos.Despues va a tomar el historial de entrenamiento que se guardo en la variable history, y despues mostrara una gráfica de línea que muestra cómo evolucionó la pérdida del modelo en cuanto a las epocas que agregamos anteriormente en las lienas de codigo pasadas.

```python
loss, accuracy = modelo.evaluate(X, y)
print(f'Loss: {loss}, Accuracy: {accuracy}')
```

Nos mostrara con números y esto nos mostrara qué tan bien funciona el modelo al final. El valor de Loss (Pérdida) bajo es generalmente bueno y el valor de Accuracy (Exactitud).


```python
pred = (modelo.predict(X) > 0.5).astype(int)
for i, (inp, p) in enumerate(zip(X, pred)):
    print(f"{inp} -> {p[0]}")
```

Este ultimo codigo nos muestra si el modelo esta entrenado para tomar una decisión binaria (0 y 1) para cada entrada, y muestra la decision del modelo si en este caso aprendio o no.
