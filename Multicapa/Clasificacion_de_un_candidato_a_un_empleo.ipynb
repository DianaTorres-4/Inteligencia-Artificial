```python
import numpy as np
import pandas as pd

# Generar dataset sint茅tico con diferentes rangos por habilidad
n_muestras = 5000
# Habilidades t茅cnicas: 70-100
tec = np.random.randint(70, 101, (n_muestras, 6))
# Habilidades matem谩ticas: 60-90
mat = np.random.randint(60, 91, (n_muestras, 6))
# Habilidades psicol贸gicas: 50-85
psi = np.random.randint(50, 86, (n_muestras, 6))
# Nombres de columnas
columnas = [
    'python', 'sql', 'visualizacion_datos', 'excel', 'manejo_apis', 'nube',
    'estadistica', 'probabilidad', 'algebra_lineal', 'calculo', 'optimizacion', 'modelado',
    'trabajo_equipo', 'comunicacion', 'creatividad', 'adaptabilidad', 'liderazgo', 'resolucion_problemas'
]
# Crear DataFrame
df = pd.DataFrame(np.concatenate([tec, mat, psi], axis=1), columns=columnas)
```

primero importaremos la libreria de pandas despues genraremos un dataser sintetico que los trabajaremos con diferentes rangos por habilidad que seran clasificados en 3 habilidades cada clasificacion debe de tener un rango de puntaje en este caso la primer habilidad tendra un rango de 70-100 la segunda habilidad es de 60-90 y la tercera habilidad sera de 50-85 y pediremos que klo genere en un excel con cada una de las columnas que pedimos que es lo que se encuentera en rojo del codigo y de este modo pedimos crear un dataframe

```python
# Calcular promedio por grupo y promedio total
prom_tecnologicas = df[['python', 'sql', 'visualizacion_datos', 'excel', 'manejo_apis', 'nube']].mean(axis=1)
prom_matematicas  = df[['estadistica', 'probabilidad', 'algebra_lineal', 'calculo', 'optimizacion', 'modelado']].mean(axis=1)
prom_psicologicas = df[['trabajo_equipo', 'comunicacion', 'creatividad', 'adaptabilidad', 'liderazgo', 'resolucion_problemas']].mean(axis=1)
# Promedio total = promedio de los 3 promedios
prom_total = (prom_tecnologicas + prom_matematicas + prom_psicologicas) / 3
# Etiqueta: 1 = apto si promedio total > 80
y = (prom_total > 75).astype(int)
# one-hot encoding
y = pd.get_dummies(y).values
```

despues pediremos calcular el promedio por grupo y promedio total de cada una de las habilidades y despues calcular el promedio total de las 3 habilidades

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df, y, test_size=0.2, random_state=42
)

# Escalamiento
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

en las lineas de codigo usaremos una libreria que nos ayudara a dividir el conjunto de datos para una evaluaci贸n justa y luego los estandariza para optimizar el rendimiento de la red neuronal que ya hemos compilado y entrenenado

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
modelo = Sequential([
    Dense(64, input_shape=(X_train.shape[1],), activation='relu'),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(2, activation='softmax')
])

adam = Adam(learning_rate=0.001)
modelo.compile(optimizer=adam, loss='categorical_crossentropy', metrics=['accuracy'])
```

Le pedimos al modelo que en la primera capa tenga 64 neuronas, 32 en la segunda capa , 16 en la tercera y dos neuronas en la salida utilizando Adam para actualizar los pesos

```python
history = modelo.fit(
    X_train, y_train,
    epochs=50,
    batch_size=8,
    validation_data=(X_test, y_test),
    verbose=1
)
```

en estas lineas de c贸digo le estamos pidiendo al modelo que entrene usando 50 ciclos sobre los datos de X_train e y_train, actualizando sus par谩metros cada 8 muestras y lo que valla entrenando en cada epoca debe de guardarlo

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='P茅rdida de entrenamiento')
plt.plot(history.history['val_loss'], label='P茅rdida de validaci贸n')
plt.xlabel('pocas')
plt.ylabel('P茅rdida')
plt.legend()
plt.title('Evoluci贸n de la funci贸n de p茅rdida')
plt.show()

loss, acc = modelo.evaluate(X_test, y_test)
print(f"\n Loss: {loss:.4f}  |  Accuracy: {acc:.4f}")
```

pedimos que nos muestre una garfica para ver si el modelo aprendio o no y nos muestre tood de una manera mas visual y facil de entender


```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, recall_score

pred = modelo.predict(X_test)
y_pred = np.argmax(pred, axis=1)
y_true = np.argmax(y_test, axis=1)

cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['No Apto', 'Apto'])
disp.plot(cmap='Blues')
plt.show()

sensitivity = recall_score(y_true, y_pred, average=None)
print("\nSensitivity (Recall) por clase:")
print(f"No Apto: {sensitivity[0]:.2f}")
print(f"Apto: {sensitivity[1]:.2f}")
```

pedimos que los datos de la muestre los coloque en una matriz y asi podemos visualizar y entender un poco mejor el modelo y ver  cuantas personas estan aptas y cuantas no segun la matriz el resultado que nos muestra

```python
nuevo_candidato = np.array([[90, 88, 85, 87, 84, 89,   # t茅cnicas
                             82, 86, 85, 83, 87, 80,   # matem谩ticas
                             70, 75, 80, 72, 78, 74]]) # psicol贸gicas

nuevo_candidato_scaled = scaler.transform(nuevo_candidato)
pred_nuevo = modelo.predict(nuevo_candidato_scaled)
clase_predicha = np.argmax(pred_nuevo)

print("Evaluaci贸n de nuevo candidato:")
print(f"Resultado: {'Apto' if clase_predicha == 1 else 'No Apto'}")
```
