```python
import pandas as pd

_df = pd.read_csv("https://gist.githubusercontent.com/mevangelista-alvarado/6f4f28c00f9e683108637cb3c8d5db89/raw/1ce753283fce23eeab952fa1d660a7069bbe9104/Bitcoin_1_1_2024-6_9_2024_historical_data_coinmarketcap.csv", delimiter=';')
# Ordenamos correctamente
_df = _df.sort_index(ascending=False)
_df.head()
```

en esta linea de codigo lo podemos hacer pero desde la liga de la web

```python
'timeOpen', 'timeClose', 'timeHigh', 'timeLow', 'name', 'open', 'high',
'low', 'close', 'volume', 'marketCap', 'timestamp'
```

vamos a obtener un dataframe con las siguientes columnas que colocamos en nuestras lineas de codigo

```python
df = _df[['timeOpen', 'open', 'high', 'low', 'close']]
```

seleccionaremos la data

```python
dates = df[['close']].values
```

vamos a convertir los datos en numpy arrays

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(dates)
```

aqui utilizaremos la funcion minmaxscaler para otorgar valores entre 0 y 1

```python
window_size = 60
```

vamos a definir el tamao de la ventana de tiempo

```python
import numpy as np

def create_sequences(data, window_size):
    sequences = []
    labels = []
    for i in range(len(data) - window_size):
        sequences.append(data[i:i+window_size])
        # El precio es la primera columna
        labels.append(data[i + window_size, 0])
    return np.array(sequences), np.array(labels)

X, y = create_sequences(scaled_data, window_size)
```

se crearan la secuencia de datos tomando en cuenta el tamao de la ventanaque definimos en la linea de codigo anterior nos sirve para el numero de puntos de datos que se utilizaran para predecir

```python
split = int(len(X) * 0.8)
X_train, y_train = X[:split], y[:split]
X_test, y_test = X[split:], y[split:]
```

dividimos los datos en conjuntos de entrenamiento y conjuntos de prueba

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, SimpleRNN

model = Sequential()
model.add(SimpleRNN(units=120, return_sequences=True, input_shape=(window_size, X_train.shape[2])))
model.add(SimpleRNN(units=60, return_sequences=False))
model.add(Dense(units=30))
model.add(Dense(units=1))
```

utilizaremos el modelo RNN que sirve para procesar secuencias y luego realiza una prediccion utilizando capas que se conectarn entre si

```python
from tensorflow.keras.optimizers import Adam

learning_rate = 0.001
adam_optimizer = Adam(learning_rate=learning_rate)

# Compilar el modelo
model.compile(optimizer=adam_optimizer, loss='mean_squared_error')

# Entrenar el modelo
model.fit(X_train, y_train, batch_size=1, epochs=10)
```

hacemos el entrenamiento otorgandole epocas para ver si el modelo esta aprendiendo o no

```python
# Hacer predicciones
predictions = model.predict(X_test)

# Desescalar los datos predichos
predictions = scaler.inverse_transform(np.concatenate((predictions, np.zeros((predictions.shape[0], 3))), axis=1))[:,0]
y_test = scaler.inverse_transform(np.concatenate((y_test.reshape(-1, 1), np.zeros((y_test.shape[0], 3))), axis=1))[:,0]
```

vamos a hacer las predicciones del modelo

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_test, predictions)
rmse = np.sqrt(mean_squared_error(y_test, predictions))

print(f"MAE: {mae}")
print(f"RMSE: {rmse}")
```

esta linea de codigo nos va a mostrar las metricas las cuales ambas metricas nos ayudan a medir la longitud del error del modelo

```python
import matplotlib.pyplot as plt
import numpy as np

valid = df[split:]
# Reiniciar el ndice del DataFrame de validacin
valid = valid.reset_index(drop=True)
# Inicializar la columna de predicciones con NaN
valid['Predictions'] = np.nan

# Aadir las predicciones al DataFrame de validacin desde el punto donde inician las predicciones
valid.loc[window_size:, 'Predictions'] = predictions
dates_valid = pd.to_datetime(valid['timeOpen']).apply(lambda x: x.strftime('%Y-%m-%d')).tolist()

plt.figure(figsize=(16,8))
plt.title('Modelo RNN para Prediccin de Bitcoin')
plt.xlabel('Fecha')
plt.ylabel('Precio de Bitcoin (USD)')
plt.plot(dates_valid, valid[['close', 'Predictions']])
plt.legend(['Valor Real', 'Predicciones'], loc='lower right')
plt.xticks(rotation=90)
plt.show()
```

ahora solo pedimos que nos muestre de una manera grafica el resultado de dicho modelo.

```python
# Generar secuencias para los siguientes das
future_sequences = []
# ltima secuencia de los valores que tenemos
last_sequence = X[-1]

days = 10
for _ in range(days):
    # Predecir el siguiente valor
    next_value = model.predict(np.array([last_sequence]))[0, 0]

    # Actualizar la secuencia para la siguiente prediccin
    last_sequence = np.concatenate((last_sequence[1:], [[next_value]]), axis=0)

    # Aadir la nueva secuencia a las secuencias futuras
    future_sequences.append(last_sequence)

# Convertir las secuencias futuras a un numpy y reshape para la RNN
future_sequences = np.array(future_sequences)
future_sequences = np.reshape(future_sequences, (future_sequences.shape[0], future_sequences.shape[1], 1))

# Hacer predicciones para los siguientes das
future_predictions = model.predict(future_sequences)

# Desescalar los datos predichos para los siguientes das
future_predictions = scaler.inverse_transform(np.concatenate((future_predictions, np.zeros((future_predictions.shape[0], 3))), axis=1))[:,0]
```

en esta linea de codigo lo que haremos es calcular la prediccion del vaor de cierre de bitcoin en los proximos 115 dias

```python
# ltima fecha en los datos reales
last_date = df['timeOpen'].iloc[-1]
future_dates = pd.date_range(start=last_date, periods=days)[1:]
future_dates = future_dates.strftime('%Y-%m-%d').tolist()
```

hacemos la generacion de fechas para los siguientes dias

```python
plt.figure(figsize=(16,8))
plt.title('Predicciones de los siguientes das')
plt.xlabel('Fecha')
plt.ylabel('Precio de Bitcoin (USD)')
plt.plot(dates_valid, valid[['close', 'Predictions']], label=['real', 'Predicciones'])
plt.plot(future_dates, future_predictions[:-1], label='Predicciones')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

nos muestra graficamente la nueva prediccion que estamos haciendo
